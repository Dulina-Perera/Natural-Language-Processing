```python
import os

from google.colab import userdata
from langchain import PromptTemplate, LLMChain
from langchain_core.runnables import RunnableSequence
from langchain_huggingface import HuggingFaceEndpoint, HuggingFacePipeline
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    GPT2LMHeadModel,
    GPT2TokenizerFast,
    pipeline,
    TextGenerationPipeline,
)

hf_tkn: str = userdata.get("HF_TOKEN")
hfh_token: str = userdata.get("HFH_TOKEN")

os.environ["HUGGINGFACEHUB_API_TOKEN"] = hfh_token
```

## Hugging Face Endpoints

```python
repo_id: str = "mistralai/Mistral-7B-Instruct-v0.2"

llm: HuggingFaceEndpoint = HuggingFaceEndpoint(
    repo_id=repo_id, max_length=128, temperature=0.7, token=hfh_token
)

print("\n")
llm.invoke("What is machine learning?")
```

```python
repo_id: str = "mistralai/Mistral-7B-Instruct-v0.3"

llm = HuggingFaceEndpoint(
    repo_id=repo_id, max_length=128, temperature=0.7, token=hfh_token
)

print("\n")
llm.invoke("What is Genertaive AI?")
```

```python
question: str = "Who won the ICC T20 World Cup in the year 2014?"
template: str = """Question: {question} Answer: Let's think step by step."""

prompt: PromptTemplate = PromptTemplate(template=template, input_variables=["question"])

llm_chain: LLMChain = LLMChain(llm=llm, prompt=prompt)

print("\n")
llm_chain.invoke(question)
```

## Hugging Face Pipelines

```python
model_id: str = "gpt2"

model: GPT2LMHeadModel = AutoModelForCausalLM.from_pretrained(model_id)
tokenizer: GPT2TokenizerFast = AutoTokenizer.from_pretrained(model_id)

pipe: TextGenerationPipeline = pipeline(
    "text-generation", model=model, tokenizer=tokenizer, max_new_tokens=100
)
hf: HuggingFacePipeline = HuggingFacePipeline(pipeline=pipe)

print("\n")
hf.invoke("What is machine learning")
```

```python
gpu_llm: HuggingFacePipeline = HuggingFacePipeline.from_model_id(
    model_id="gpt2",
    task="text-generation",
    device=0,
    pipeline_kwargs={"max_new_tokens": 1000},
)

prompt = PromptTemplate.from_template(template)

chain: RunnableSequence = prompt | gpu_llm

chain.invoke({"question": "What is artificial intelligence?"})
```
